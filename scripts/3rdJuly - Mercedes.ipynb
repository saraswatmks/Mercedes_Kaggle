```python
target = np.log1p(train['y'])
```

```python
train.drop('y',axis=1,inplace=True)
```

```python
x_train = train.copy()
x_test = test.copy()
```

```python
x_train= np.array(x_train)
x_test = np.array(x_test)
```

```python
y_train = target
```

```python
ntrain = train.shape[0]
ntest = test.shape[0]
```

```python
NFOLDS = 5
```

```python
from sklearn.cross_validation import KFold
```

```python
kf = KFold(ntrain, n_folds=NFOLDS, shuffle=True, random_state=42)
```

```python
from sklearn.metrics import r2_score

def rsquare_score(preds, dtrain):
    labels = dtrain.get_label()
    return 'r2_score', r2_score(labels, preds)
```

```python
class SklearnWrapper(object):
    def __init__(self, clf, seed=0, params=None):
        params['random_state'] = seed
        self.clf = clf(**params)
        
    def train(self, x_train, y_train):
        self.clf.fit(x_train, y_train)
        
    def predict(self, x):
        return self.clf.predict(x)
```

```python
class XgbWrapper(object):
    def __init__(self, seed=0, params=None):
        self.param = params
        self.param['seed'] = seed
        self.nrounds = params.pop('nrounds', 250)

    def train(self, x_train, y_train):
        dtrain = xgb.DMatrix(x_train, label=y_train)
        self.gbdt = xgb.train(self.param, dtrain, self.nrounds)

    def predict(self, x):
        return self.gbdt.predict(xgb.DMatrix(x))
```

```python
def get_oof(clf):
    oof_train = np.zeros((ntrain,))
    oof_test = np.zeros((ntest,))
    oof_test_skf = np.empty((NFOLDS, ntest))

    for i, (train_index, test_index) in enumerate(kf):
        x_tr = x_train[train_index]
        y_tr = y_train[train_index]
        x_te = x_train[test_index]

        clf.train(x_tr, y_tr)

        oof_train[test_index] = clf.predict(x_te)
        oof_test_skf[i, :] = clf.predict(x_test)

    oof_test[:] = oof_test_skf.mean(axis=0)
    return oof_train.reshape(-1, 1), oof_test.reshape(-1, 1)
```

```python
SEED = 0
```

```python
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor
import xgboost as xgb
```

```python
et_params = {
    'n_estimators': 100,
    'max_features': 0.4,
    'max_depth': 6,
    'min_samples_split': 7,
}

rf_params = {
    'n_estimators': 100,
    'max_features': 0.2,
    'max_depth': 8,
    'min_samples_split': 7,
}

xgb_params = {
    'seed': 0,
    'colsample_bytree': 0.7,
    'subsample': 0.7,
    'learning_rate': 0.025,
    'objective': 'reg:linear',
    'max_depth': 7,
    'min_child_weight': 0,
    'nrounds': 500
}
```

```python
xg = XgbWrapper(seed=SEED, params=xgb_params)
et = SklearnWrapper(clf=ExtraTreesRegressor, seed=SEED, params=et_params)
rf = SklearnWrapper(clf=RandomForestRegressor, seed=SEED, params=rf_params)
```

```python
xg_oof_train, xg_oof_test = get_oof(xg)
```

```python
len(xg_oof_train), len(xg_oof_test)
```

```python
et_oof_train, et_oof_test = get_oof(et)
```

```python
rf_oof_train, rf_oof_test = get_oof(rf)
```

```python
print("XG-CV: {}".format(r2_score(y_train, xg_oof_train)))
print("ET-CV: {}".format(r2_score(y_train, et_oof_train)))
print("RF-CV: {}".format(r2_score(y_train, rf_oof_train)))
```

```python
x_train = np.concatenate((xg_oof_train, et_oof_train, rf_oof_train), axis=1)
x_test = np.concatenate((xg_oof_test, et_oof_test, rf_oof_test), axis=1)
```

```python
print("{},{}".format(x_train.shape, x_test.shape))

dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test)
```

```python
xgb_params = {
    'seed': 0,
    'colsample_bytree': 0.8,
    'subsample': 0.6,
    'eta': 0.025,
    'objective': 'reg:linear',
    'max_depth': 6,
    'min_child_weight': 0,
}
```

```python
res = xgb.cv(xgb_params, dtrain, num_boost_round=1000, nfold=4, seed=SEED, stratified=False,
             early_stopping_rounds=40, verbose_eval=10, show_stdv=True,feval = rsquare_score,maximize = True)
```

```python
best_nrounds = res.shape[0] - 1
best_nrounds
```

```python
cv_mean = res.iloc[-1, 0]
cv_std = res.iloc[-1, 1]
cv_mean, cv_std
```

```python
print('Ensemble-CV: {0}+{1}'.format(cv_mean, cv_std))
```

```python
gbdt = xgb.train(xgb_params, dtrain, best_nrounds)
```

```python
submission = pd.read_csv("sample_submission.csv")
submission.iloc[:, 1] = np.exp(gbdt.predict(dtest))-1
```
